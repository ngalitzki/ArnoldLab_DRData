```python
import subprocess as sp
import os, time, sys
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

path = '/home/polarbear/PB2B_Backend/Cooldown_Runs/20170131-04b/Data/'
```

```python
fileNames = os.listdir(path)
files = [path + s for s in fileNames]
files = sorted(files, key = os.path.getctime)
files
```

```python
data = np.array([])
for f in files:
    if f[-4:] == '.txt' and f[-7:] != 'RAW.txt' and f[-10:] != 'MASTER.txt':
        print 'loading from: ', f
        newData = np.loadtxt(f, skiprows = 3, delimiter = ',')
        print 'data shape: ', np.shape(newData)
        if len(newData) != 0:
            if len(data) == 0:
                data = newData
            else:
                data = np.vstack((data, newData))
```

```python
plt.figure(figsize = (12,8))
for i in range(1, np.shape(data)[1], 1):
    if i != 6 or i != 7 or i != 8:
        plt.plot(data[:, 0], data[:, i])
    plt.ylim((0, 350))
```

```python
rawdata = np.array([])
for f in files:
    if f[-7:] == 'RAW.txt':
        print 'loading from: ', f
        newData = np.loadtxt(f, skiprows = 4, delimiter = ',')
        print 'data shape: ', np.shape(newData)
        if len(newData) != 0:
            if len(rawdata) == 0:
                rawdata = newData
            else:
                rawdata = np.vstack((rawdata, newData))
```

```python
with open(files[1], 'r') as f:
    f.seek(66)
    column_headers = f.readline().strip('\n')
    
headerStr = 'Date: ' + time.strftime("%Y/%m/%d") + ', Time: ' + \
            time.strftime("%H:%M:%S") + '\n________________________________\n' + column_headers

masterFileName = time.strftime("%Y%m%d-%H%M%S" + '_MASTER.txt')

np.savetxt(path + masterFileName, data, fmt = '%.16e', delimiter = ',', newline = '\n', header = headerStr, comments = '')
```
